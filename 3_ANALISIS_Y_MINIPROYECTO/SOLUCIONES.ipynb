### 游깯 **4. Visualizaci칩n avanzada con Seaborn**

**Objetivo**: Visualizar datos categ칩ricos.

**Enunciado**:
1. Carga el dataset `tips` de Seaborn.
2. Realiza un gr치fico de barras mostrando la propina promedio (tip) por d칤a.
3. Crea un gr치fico de caja para analizar la distribuci칩n de propinas (tip) por d칤a.

**Mini-gu칤a**:
- Usa `sns.load_dataset('tips')`.
- M칠todos 칰tiles: `sns.barplot()`, `sns.boxplot()`, `plt.show()`.

```python
df = sns.load_dataset('tips')
print(df)

df_new =df.groupby(by=['day'])['tip'].mean().reset_index()
df_new.columns = ['days', 'mean_tips']
print(df_new)

plt.figure(figsize=(10,6))
sns.barplot(x=df_new['days'], y=df_new['mean_tips'])
plt.title('Media de propina por d칤a')
plt.xlabel('Dias')
plt.ylabel('Media de propina')

plt.figure(figsize=(10,6))
sns.boxplot(data=df, x='day', y='tip')
plt.title('Distribuci칩n de propinas por dia')
plt.xlabel('Dias')
plt.ylabel('Propinas')
```

```python
df = sns.load_dataset('tips')
print(df)

plt.figure(figsize=(10,6))
sns.barplot(x=df['day'], y=df['tip'], estimator='mean', palette='pastel')
plt.title('Media de propina por d칤a')
plt.xlabel('Dias')
plt.ylabel('Media de propina')

plt.figure(figsize=(10,6))
sns.boxplot(data=df, x='day', y='tip')
```

### 游늴 **5. An치lisis estad칤stico con Pandas**

**Objetivo**: Realizar an치lisis estad칤stico descriptivo.

**Enunciado**:
1. Carga el dataset `titanic` de Seaborn.
2. Encuentra la media, mediana, y moda de la edad de los pasajeros.
3. Encuentra cu치ntos pasajeros sobrevivieron y cu치ntos no.
4. Calcula el porcentaje de supervivencia por clase (1춹, 2춹, 3춹 clase).

**Mini-gu칤a**:
- M칠todos 칰tiles: `df.describe()`, `df['column'].mean()`, `df['column'].value_counts()`.

```python
df = sns.load_dataset('titanic')
#print(df)
print(df.describe())
media_edad = df['age'].mean()
mediana_edad = df['age'].median()
moda_edad = df['age'].mode()

print(f"Media de edad: {media_edad:.2f}\nMediana de edad: {mediana_edad:.2f}\nModa de la edad: {moda_edad[0]}")

no_sobrevive = df[df['survived'] ==0]['survived'].count()
sobrevive = (df[df['survived'] ==1])['survived'].count()
print(f"Sobreviven: {sobrevive}\nNo sobreviven: {no_sobrevive}")

porcentaje_viven = (sobrevive * 100)/len(df)
porcentaje_no_viven = (no_sobrevive *100)/len(df)

print(f"Sobreviven: {porcentaje_viven}%\nNo sobreviven: {porcentaje_no_viven}%")
```
