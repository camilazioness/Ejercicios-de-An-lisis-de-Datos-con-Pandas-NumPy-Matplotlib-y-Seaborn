## Ejercicio 5: Scatter Plot con Matplotlib 游릭
### Objetivo:
Visualizar la relaci칩n entre dos variables.

1. Usa el dataset `iris` de Seaborn.
2. Crea un gr치fico de dispersi칩n (`scatter plot`) para visualizar la relaci칩n entre el ancho del s칠palo (`sepal_width`) y la longitud del s칠palo (`sepal_length`).
3. A침ade colores diferentes seg칰n la especie de flor.

### Mini-gu칤a:
- M칠todos 칰tiles: `plt.scatter()`, `plt.color()`.

```python
# 1. Usa el dataset `iris` de Seaborn.
df = sns.load_dataset('iris')
print(df)

# 3. A침ade colores diferentes seg칰n la especie de flor.
def species_color(species):
    if species == 'setosa': 
        color = 'green'
    elif species == 'versicolor':
        color = 'red'
    elif species == 'virginica':
        color = 'orange'
    return color

df['color_x_specie'] = df['species'].apply(species_color)

# 2. Crea un gr치fico de dispersi칩n (`scatter plot`) para visualizar la relaci칩n entre el ancho del s칠palo (`sepal_width`) y la longitud del s칠palo (`sepal_length`).
plt.scatter(x=df['sepal_width'], y=df['sepal_length'], c=df['color_x_specie'])
plt.title('Relacion entre ancho y largo del sepalo')
plt.xlabel('Ancho del sepalo')
plt.ylabel('Largo del sepalo')


print(df['species'].unique())
```

## Ejercicio 7: Gr치fico de L칤nea con Media M칩vil 游늵
### Objetivo:
Visualizar series temporales.

1. Genera una serie temporal de 100 d칤as con valores aleatorios.
2. Calcula la media m칩vil de los 칰ltimos 10 d칤as y gr치ficala junto a la serie original.
3. Usa diferentes colores para la serie original y la media m칩vil.

```python
# 1. Genera una serie temporal de 100 d칤as con valores aleatorios.
df = pd.DataFrame(pd.date_range(start="1/1/2024", periods=100), columns=['date'])
df['values'] = np.random.randint(101, size=100)
print(df)


# 2. Calcula la media m칩vil de los 칰ltimos 10 d칤as y gr치ficala junto a la serie original (Usa diferentes colores para la serie original y la media m칩vil).
df['mean_10_days'] = df['values'].rolling(window=10).mean()
print(df)
plt.figure(figsize=(10, 6))
plt.plot(df['date'], df['values'], color='green', linestyle='--', linewidth=0.9, label='Valores originales')
plt.plot(df['date'], df['mean_10_days'], color='red', label='Media movil 10 dias')


# 3. A침ade titulo y etiquetas
plt.title('Camparativa de valores originales y media de 10 dias')
plt.xlabel('D칤as')
plt.ylabel('Valores')
plt.legend()
```

## Ejercicio 9: Visualizaci칩n de Datos Temporales con An치lisis Estad칤stico 游댃
### Objetivo:
Trabajar con series temporales y calcular estad칤sticas.

1. Genera una serie temporal de 200 d칤as con valores aleatorios.
2. Calcula la media y la desviaci칩n est치ndar de la serie temporal.
3. Visualiza la serie junto con las bandas de desviaci칩n est치ndar (es decir, la media 췀 1 desviaci칩n est치ndar).

```python
# 1. Serie temporal 
df = pd.DataFrame({'dates': pd.date_range(start='1-1-2024', periods=200), 'valores': np.random.randint(101, size=200)})
print(df)

# 2. C치lculos de meida y desviaci칩n estandar
media = df['valores'].mean()
desiacion_estandar = df['valores'].std()
print(media,desiacion_estandar)

# Crear gr치fica
plt.plot(figsize=(10,6))
plt.plot(df['dates'], df['valores'], label='Serie temporal', linewidth=0.7) # Gr치fica de la serie
plt.axhline(media, color='red' , label = 'Media') # A침adir linea de la media
plt.axhline(media + desiacion_estandar, color='orange', label= 'media + std') # media + std
plt.axhline(media - desiacion_estandar, color= 'green', label= 'media - std') # media -std
plt.legend()
```
