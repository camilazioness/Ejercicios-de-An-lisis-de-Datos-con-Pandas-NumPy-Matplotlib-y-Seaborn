### 游늴 **5. An치lisis estad칤stico con Pandas**

**Objetivo**: Realizar an치lisis estad칤stico descriptivo.

**Enunciado**:
1. Carga el dataset `titanic` de Seaborn.
2. Encuentra la media, mediana, y moda de la edad de los pasajeros.
3. Encuentra cu치ntos pasajeros sobrevivieron y cu치ntos no.
4. Calcula el porcentaje de supervivencia por clase (1춹, 2춹, 3춹 clase).

```python
# Cargar dataset
df = sns.load_dataset('titanic')
#print(df)
#print(df.describe())

# C치lculo estad칤stico
media_edad = df['age'].mean()
mediana_edad = df['age'].median()
moda_edad = df['age'].mode()

print(f"Media de edad: {media_edad:.2f}\nMediana de edad: {mediana_edad:.2f}\nModa de la edad: {moda_edad[0]}")

# Filtrar los valores de la columnas y conatar cuantas filas hay (.count())
sobrevive = df[df['survived'] ==1]['survived'].count()
no_sobrevive = df[df['survived'] ==0]['survived'].count()
print(f"Sobreviven: {sobrevive}\nNo sobreviven: {no_sobrevive}")

# Calcular los porcentajes 
porcentaje_sobrevive = (sobrevive * 100)/len(df)
porcentaje_no_sobrevive = (no_sobrevive *100)/len(df)

# Imprimir
print(f"Sobreviven: {porcentaje_sobrevive:.2f} %\nNo sobreviven: {porcentaje_no_sobrevive:.2f} %")
```

### 游늰 **9. An치lisis de datos temporales**

**Objetivo**: Trabajar con series temporales.

**Enunciado**:
1. Crea una serie temporal de 100 d칤as con valores aleatorios entre 50 y 100.
2. Calcula la media m칩vil con una ventana de 10 d칤as.
3. Grafica la serie temporal original y la media m칩vil en la misma gr치fica.

```python
# Dataframe con fecha sy valores aleatorios asociados
df = pd.DataFrame(pd.date_range(start='01-01-2024', periods=100), columns=['days'])
df['values'] = np.random.randint(50, 101, size=100)

# Calcular la media movil 
df['mean_10_days'] = df['values'].rolling(window=10).mean()
print(df)

# Graficar la serie temporal
plt.figure(figsize=(10,6))
sns.lineplot(df, x='days', y='values', linestyle='--', linewidth=0.7, label= 'Valores originales')
sns.lineplot(df, x='days', y='mean_10_days', color='red', linewidth=0.9, label= 'Media de 10 dias')

plt.title('Camparativa de valores originales y media de 10 dias')
plt.xlabel('D칤as')
plt.ylabel('Valores')
plt.legend()
```

### 游늳 **10. Aplicar t칠cnicas de Estad칤stica descriptiva**

**Objetivo**: Trabajar con distribuci칩n de frecuencias y medidas de tendencia central.

**Enunciado**:
1. Carga el dataset `diamonds` de Seaborn.
2. Calcula la frecuencia de los diferentes cortes de diamantes.
3. Encuentra la media, mediana y moda de los precios de los diamantes.
4. Visualiza la distribuci칩n del precio con un gr치fico de densidad.


```python
# Cargar dataset
df = sns.load_dataset('diamonds')
#print(df)

# Calcular la frecuencia de los diferentes cortes de diamantes
count_cuts = df['cut'].value_counts()
print(count_cuts)

# C치lculo estadistico
mean_price = df['price'].mean()
median_price = df['price'].median()
mode_price = df['price'].mode()[0]

print(f"media = {mean_price}, mediana = {median_price}, moda = {mode_price}")

# Graficar la distribuci칩n del precio
plt.figure(figsize=(10,6))
sns.kdeplot(df['price'], fill=True, label='Densidad de precios')
plt.axvline(mean_price, color= 'green', linestyle = '--', label= 'Media de precio')
plt.axvline(median_price, color= 'orange', linestyle = '--', label= 'Mediana de precio')
plt.axvline(mode_price, color= 'red', linestyle = '--', label= 'Moda de precio')
plt.legend()
```

### 游빌 **11. Agrupaciones y operaciones complejas**

**Objetivo**: Trabajar con agrupaciones y operaciones avanzadas.

**Enunciado**:
1. Carga el dataset `titanic` de Seaborn.
2. Agrupa los datos por clase y sexo, y calcula la media de la edad y el precio del billete.
3. Crea un gr치fico de barras que muestre la media del precio del billete por clase.


```python
# Cargar dataset
df = sns.load_dataset('titanic')
#print(df)

# Agrupar datos mediante una fucion
def calcular_media(df, col_agrupada, col_media):
    df_group = df.groupby(by=col_agrupada, observed=True)[col_media].mean()
    return df_group

# Llamar a la funcion de agrupacion
age_clase = calcular_media(df, ['class'], ['age'])
age_sex = calcular_media(df, ['sex'], ['age'])
price_class = calcular_media(df, ['class'], ['fare'])
price_sex = calcular_media(df, ['sex'], ['fare'])

# Imprimir resultados
#print(age_clase)
#print(age_sex)
print(price_class)
#print(price_sex)

# Crear gr치fico 
plt.figure(figsize=(10,6))
sns.barplot(df, x='class', y='fare', estimator='mean')
plt.title('Media de precio por clase')
plt.xlabel('Clases')
plt.ylabel('Media del precio del billete')
```
