
## Ejercicio 3: Histograma con Matplotlib 🛠️
### Objetivo:
Visualizar la distribución de datos numéricos.

1. Genera 1000 números aleatorios con distribución normal (media = 0, desviación estándar = 1).
2. Crea un histograma usando Matplotlib para visualizar la distribución.
3. Ajusta el número de "bins" del histograma para que la visualización sea clara.

### Mini-guía:
- Usa `np.random.normal()` para generar los números.
- Métodos útiles: `plt.hist()`, `plt.grid()`, `plt.show()`.

```python
# 1. Genera 1000 números aleatorios con distribución normal (media = 0, desviación estándar = 1).
media = 0
std = 1
numeros_aleatorios = np.random.normal(media, std, 1000)

# 2. Crea un histograma usando Matplotlib para visualizar la distribución.
plt.figure(figsize=(10,6))
plt.hist(numeros_aleatorios)
plt.title('Distribución')
plt.xlabel('Numeros aleatorios')
plt.ylabel('Frecuencia')

# 3. Ajusta el número de "bins" del histograma para que la visualización sea clara.
plt.figure(figsize=(10,6))
plt.hist(numeros_aleatorios, bins=50, color='orange')
plt.title('Distribución')
plt.xlabel('Numeros aleatorios')
plt.ylabel('Frecuencia')
plt.grid()
```

## Ejercicio 5: Scatter Plot con Matplotlib 🟢
### Objetivo:
Visualizar la relación entre dos variables.

1. Usa el dataset `iris` de Seaborn.
2. Crea un gráfico de dispersión (`scatter plot`) para visualizar la relación entre el ancho del sépalo (`sepal_width`) y la longitud del sépalo (`sepal_length`).
3. Añade colores diferentes según la especie de flor.

### Mini-guía:
- Métodos útiles: `plt.scatter()`, `plt.color()`.

```python
# 1. Usa el dataset `iris` de Seaborn.
df = sns.load_dataset('iris')
print(df)

# 3. Añade colores diferentes según la especie de flor.
def species_color(species):
    if species == 'setosa': 
        color = 'green'
    elif species == 'versicolor':
        color = 'red'
    elif species == 'virginica':
        color = 'orange'
    return color

df['color_x_specie'] = df['species'].apply(species_color)

# 2. Crea un gráfico de dispersión (`scatter plot`) para visualizar la relación entre el ancho del sépalo (`sepal_width`) y la longitud del sépalo (`sepal_length`).
plt.figure(figsize=(10,6))
plt.scatter(x=df['sepal_width'], y=df['sepal_length'], c=df['color_x_specie'])
plt.title('Relacion entre ancho y largo del sepalo')
plt.xlabel('Ancho del sepalo')
plt.ylabel('Largo del sepalo')
```

## Ejercicio 6: Heatmap con Seaborn 🔥
### Objetivo:
Visualizar la correlación entre variables numéricas.

1. Usa el dataset `iris` de Seaborn.
2. Calcula la matriz de correlación de las variables.
3. Crea un heatmap para visualizar la matriz de correlación.

```python
#1. Usa el dataset `iris` de Seaborn.
df = sns.load_dataset('iris')

#2. Calcula la matriz de correlación de las variables. (Selecciono los tipos de variable que quiero que esté en mi matriz)
corr_matrix = df.select_dtypes(include=['float64', 'int64']).corr()
print(corr_matrix)

#3. Crea un heatmap para visualizar la matriz de correlación.
plt.figure(figsize=(10,6))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm')
```

## Ejercicio 7: Gráfico de Línea con Media Móvil 📊
### Objetivo:
Visualizar series temporales.

1. Genera una serie temporal de 100 días con valores aleatorios.
2. Calcula la media móvil de los últimos 10 días y gráficala junto a la serie original.
3. Usa diferentes colores para la serie original y la media móvil.

```python
# 1. Genera una serie temporal de 100 días con valores aleatorios.
df = pd.DataFrame(pd.date_range(start="1/1/2024", periods=100), columns=['date'])
df['values'] = np.random.randint(101, size=100)
#print(df)


# 2. Calcula la media móvil de los últimos 10 días y gráficala junto a la serie original (Usa diferentes colores para la serie original y la media móvil).
df['mean_10_days'] = df['values'].rolling(window=10).mean()
#print(df)
plt.figure(figsize=(10, 6))
plt.plot(df['date'], df['values'], color='green', linestyle='--', linewidth=0.9, label='Valores originales')
plt.plot(df['date'], df['mean_10_days'], color='red', label='Media movil 10 dias')


# 3. Añade titulo y etiquetas
plt.title('Camparativa de valores originales y media de 10 dias')
plt.xlabel('Días')
plt.ylabel('Valores')
plt.legend()
```

## Ejercicio 9: Visualización de Datos Temporales con Análisis Estadístico 🔄
### Objetivo:
Trabajar con series temporales y calcular estadísticas.

1. Genera una serie temporal de 200 días con valores aleatorios.
2. Calcula la media y la desviación estándar de la serie temporal.
3. Visualiza la serie junto con las bandas de desviación estándar (es decir, la media ± 1 desviación estándar).

```python
# 1. Serie temporal 
df = pd.DataFrame({'dates': pd.date_range(start='1-1-2024', periods=200), 'valores': np.random.randint(101, size=200)})
#print(df)

# 2. Cálculos de meida y desviación estandar
media = df['valores'].mean()
desiacion_estandar = df['valores'].std()
#print(media,desiacion_estandar)

# Crear gráfica
plt.figure(figsize=(10,6))
plt.plot(df['dates'], df['valores'], label='Serie temporal', linewidth=0.7) # Gráfica de la serie
plt.axhline(media, color='red' , label = 'Media') # Añadir linea de la media
plt.axhline(media + desiacion_estandar, color='orange', label= 'media + std') # media + std
plt.axhline(media - desiacion_estandar, color= 'green', label= 'media - std') # media -std
plt.legend()
```
